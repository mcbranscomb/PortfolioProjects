# 911 Calls Data Project

```python
# Importing needed libraries

import numpy as np
import pandas as pd
```

```python
# Importing plot and visualization libraries

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
# Returning dataset info

df.info()
```

```python
# Populating dataset

df.head()
```

```python
# Returning zipcodes with highest 911 calls received

df['zip'].value_counts().head(5)
```

```python
# Returning townships with highest 911 calls received

df['twp'].value_counts().head(5)
```

```python
# Exploring number of unique title codes

df['title'].nunique()
```

```python
# Creating new column from titles column using lambda expression. New 'Reason' column shows
# nature of phone call: EMS/Fire/Traffic

df['Reason'] = df['title'].apply(lambda reason:reason.split(':')[0])    
```

```python
# Returning most common reasons 911 was dialed

df['Reason'].value_counts()
```

```python
# Returning countplot of 911 calls by 'Reason'

sns.countplot(x='Reason',data=df,palette='viridis')
```

```python
# Returning data time in 'timeStamp' column

type(df['timeStamp'].iloc[0])
```

```python
# Converting str data type to DataTime objects

df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
# Creating 'Hour', 'Month', 'Day of Week' columns by calling attributes of 'timeStamp' column

df['Hour'] = df['timeStamp'].apply(lambda time:time.hour)
df['Month'] = df['timeStamp'].apply(lambda time:time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
# Dictionary to map string names to day of week

dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
# Returning countplot of the Day of Week

sns.countplot(x='Day of Week',data=df,hue='Reason',palette='viridis')
```

```python
# Returning countplot for Month

sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')
```

```python
# Returning Month groupby object to search for missing months in the above countplot. Using count() for aggregation

byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
# Plotting count of calls per month

byMonth['twp'].plot()
```

```python
# Linear plot of calls per month

sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
# Creating new Date column using lambda function to pull date from timeStamp column

df['Date']=df['timeStamp'].apply(lambda t: t.date())
```

```python
# Aggregating date column values with count() and plotting

df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
# Plotting Traffic calls by month

df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
# Plotting fire calls by month

df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
# Plotting EMS calls by month

df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

```python
# Restructuring dataframe: column hours and Day of Week index. Reason column unstacked for heatmap visual

dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
# Heatmap visual

plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

```python
# Clustermap visual

sns.clustermap(dayHour,cmap='viridis')
```

```python
# Dataframe build by month. Reason column unstacked for visuals

dayMonth = df.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
# Heatmap of month dataframe

plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis')
```

```python
# Clustermap of month dataframe

sns.clustermap(dayMonth,cmap='viridis')
```
